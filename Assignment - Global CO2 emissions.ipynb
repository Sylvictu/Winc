```python
import pandas as pd
wiki_read = pd.read_html('https://en.wikipedia.org/wiki/List_of_countries_by_carbon_dioxide_emissions')
data = wiki_read[1]
```

```python
data.head(10)
```

```python
data.columns = ['country',1990,2005,2017,'2017 %','2017 vs 1990', 'per land', 'per capita', 'total inc','total exl']
```

```python
data.drop(axis=1,columns=data.columns[4:],inplace=True)
```

```python
data
```

```python
data.nlargest(5, 2017)
```

```python
data = data.drop(labels=[0, 1 , 2, 67],axis = 0,)
```

```python
data_large = data.nlargest(5, 2017)
```

```python
data_large
```

```python
import matplotlib.pyplot as plt
```

```python
fig,ax = plt.subplots()
years = data_large.columns[1:]

for index, row in data_large.iterrows():
  plt.plot(years, row [1:], label=row)

plt.title("fossil CO2 emissions in Mt CO2")
plt.ylabel("ammount in 1.000")
plt.xlabel("Years")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()

plt.show()
```

```python
data
```

```python
data['1990'] = data[1990] / data[1990] *100
```

```python
data['2005'] = data[2005] / data[1990] *100
```

```python
data['2017'] = data[2017] / data[1990] *100
```

```python
data.head()
```

```python
data['difference_over_entire_perioid'] = data['2017'] + data['2005'] + data['2017'] 
```

```python
data.head()
```

```python
data_percent = data[data[1990] > 5]  
```

```python
data_percent.head(100)
```

```python
high_difference = data_percent.nlargest(3, "difference_over_entire_perioid")
```

```python
low_difference = data_percent.nsmallest(3, "difference_over_entire_perioid")
```

```python
high_difference.head()
```

```python
low_difference.head()
```

```python
high_difference.drop([1990,2005,2017,"difference_over_entire_perioid"],axis = 1, inplace =True)
```

```python
low_difference.drop([1990,2005,2017,"difference_over_entire_perioid"],axis = 1, inplace =True)
```

```python
high_difference
```

```python
low_difference
```

```python
fig,ax = plt.subplots()
years = high_difference.columns[1:]

for index, row in high_difference.iterrows():
  plt.plot(years, row [1:], label=row)

plt.title("Increase in usage")
plt.ylabel("increase in percentage")
plt.xlabel("Years")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()

plt.show()
```

```python
fig,ax = plt.subplots()
years = low_difference.columns[1:]

for index, row in low_difference.iterrows():
  plt.plot(years, row [1:], label=row)

plt.title("Increase in usage")
plt.ylabel("Decreace in percentage")
plt.xlabel("Years")

plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
plt.grid()

plt.show()
```
